```r
# a synthetic example
dose <- c(20, 30, 40, 45, 60)
drugA <- c(16, 20, 27, 40, 60)
drugB <- c(15, 18, 25, 31, 40)
opar <- par(no.readonly=T)
par(pin=c(2,3))
par(lwd=2, cex=1.5)
par(cex.axis=.75, font.axis=3)
plot(dose, drugA, type='b', pch=19, lty=2, col='red')
plot(dose, drugB, type='b', pch=23, 
     lty=6, col='blue', bg='green')
par(opar)
```

### add text, user-defined axis and graph

```r
plot(dose, drugA, type='b',
    col='red', lty=2, lwd=2, pch=2,
    main='Clinical Trails for Drug A',
    sub='This is a hypothetical data',
    xlab='Dosage', ylab='Drug Resbonse',
    xlim=c(0, 60), ylim=c(0, 70))
```

```r
# notice that some 'plot' function will automatically 
#      add a title and a legend, which can be removed by
#                parameter: ann=F
```

### function title()

```r
opar <- par(no.readonly=T)
plot(dose, drugA, type='b',
    col='red', lty=2, lwd=2, pch=2,
    xlim=c(0, 60), ylim=c(0, 70),
    ann=F)         # ann=F to remove the primitive label
title(main='Clinical Trails for Drug A',
     sub='This is a hypothetical data',
     col.main = 'red', col.sub='blue',
     xlab='Dosage', ylab='Drug Resbonse',
     col.lab='green', cex.lab=0.75)
par(opar)
```

#### notice that function title() is only valid after function plot()

### function axis()

```r
# axis parameters
# side:   which side to draw an axis (1=down,2=left,3,4)
# at:     the position of a scale
# labels: default to the value of 'at'
# pos:    the position of the cross
# lty:    line type
# col:    the color of line and scale
# las:    ??
# tck:    -x means outside the graph
#         +x means inside the graph
#         0  means no scale
#         1  means nest scale
```

## a synthetic example for self-defining axis

```r
# generate data
x <- c(1:10)
y <- x
z <- 10/x

opar <- par(no.readonly=T)

par(mar=c(5, 4, 4, 8) + 0.1)  # enlarge the border

plot(x, y, type='b',
    pch=21, col='red',
    yaxt='n', lty=3, ann=F)

lines(x, z, type='b', pch=22, col='blue', lty=2)

# define your own axis
axis(2, at=x, labels=x, col.axis='red', las=2)

axis(4, at=z, labels=round(z, digits=2),
    col.axis='blue', las=2, cex.axis=0.7, tck=-0.01)

# add title and text
mtext('y=1/x', side=4, line=3, cex.lab=1, las=2, col='blue')

title('An Example of Creative Axis',
     xlab='X values',
     ylab='Y = X')

par(opar)
```

## 3.2.2 Legend

```r
# with function legend(location, title, legend, ...)
# location: locator(1) & click the position you want
#           keywords like: bottom, bottomleft, topright, etc.
#           also use inset for more accurate position
# title:    the title for the legend
# legend:   a ch-class vector of the text in legend

# for more details, please refer to help(legend)
```

```r
dose <- c(20, 30, 40, 45, 60)
drugA <- c(16, 20, 27, 40, 60)
drugB <- c(15, 18, 25, 31, 40)

opar <- par(no.readonly=T)

par(lwd=2, cex=1.5, font.lab=2)

plot(dose, drugA, type='b', pch=15, lty=1, 
     col='red', ylim=c(0, 60),
     main='Drug A vs Drug B',
     xlab='Drug Dosage', ylab='Drug Response')

lines(dose, drugB, type='b', 
      pch=17, lty=2, col='blue')

abline(h=c(30), lwd=1.5, lty=2, col='grey')

# sub-scale
# library(Hmisc)
# minor.tick(nx=3, ny=3, tick.ratio=0.5)

legend('topleft', inset=0.05, title='Drug Type', c('A','B'),
      lty=c(1, 2), pch=c(15, 17), col=c('red', 'blue'))

par(opar)
```

# 3.3 The combinition of graphs

```r
# with the parameter of par: mfrow=(nrows, ncols)
dose <- c(20, 30, 40, 45, 60)
drugA <- c(16, 20, 27, 40, 60)
drugB <- c(15, 18, 25, 31, 40)
drugC <- c(30, 33, 43, 47, 55)
drugD <- c(10, 21, 29, 38, 47)
opar <- par(no.readonly=T)
par(mfrow=c(2, 2))   # to combine as a 2*2 graph
plot(dose, drugA, type='b',
    main='DrugA', col.main='blue', col='blue')
plot(dose, drugB, type='b',
    main='DrugB', col.main='red', col='red')
plot(dose, drugC, type='b',
    main='DrugC', col.main='green', col='green')
plot(dose, drugD, type='b',
    main='DrugD', col.main='black')
par(opar)
```

```r
opar <- par(no.readonly=T)
par(mfrow=c(3, 1))
hist(drugA)
hist(drugB)
hist(drugC)
par(opar)
```

# 3.4 Boxplot

```r
# with parameter: fig=vector
opar <- par(no.readonly=T)
par(fig=c(0, 0.8, 0, 0.8))
plot(dose, drugA)

par(fig=c(0, 0.8, 0.55, 1), new=T)
boxplot(dose, horizontal = T, axes=F)
par(fig=c(0.65, 1, 0, 0.8), new=T)
boxplot(drugA, axes=F)

mtext('Enhanced Scatterplot', side=3, outer=T, line=-3)
par(opar)
```
